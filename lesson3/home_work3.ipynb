```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math

from pylab import *
from mpl_toolkits.mplot3d import Axes3D

from scipy.optimize import fsolve
```

```python
# 2.1) Напишите код на Python, реализующий расчет длины вектора, заданного его координатами
X = list(map(int, input('Введите координаты вектора через пробел: ').split(' ')))
len_vectore_X = math.sqrt(np.sum(np.power(X,2)))
print(f'Длина вектора Х: {len_vectore_X:.3f}')
```

```python
# 2.3) Напишите код на Python, реализующий построение графиков: окружности, эллипса, гипербол
fig1 = plt.figure(1)
ax1 = fig1.add_subplot(111)
fig2 = plt.figure(2)
ax2 = fig2.add_subplot(111)
fig3 = plt.figure(3)
ax3 = fig3.add_subplot(111)
R = 2
a = 10
b = 2
x_circle = np.linspace(-R, R, 201)
x_ellipse = np.linspace(-a, a, 201)
x_hyperbola = np.linspace(a, a + 5, 201)
y1_circle = []
y2_circle = []
y1_ellipse = []
y2_ellipse = []
y1_hyperbola = []
y2_hyperbola = []
for (x_c, x_e, x_h) in zip(x_circle, x_ellipse, x_hyperbola):
    y1_circle.append(math.sqrt(R**2 - x_c**2))
    y2_circle.append(-math.sqrt(R**2 - x_c**2))
    
    y1_ellipse.append(math.sqrt(b**2 * (a**2 - x_e**2) / a**2))
    y2_ellipse.append(-math.sqrt(b**2 * (a**2 - x_e**2) / a**2))
    
    y1_hyperbola.append(math.sqrt(b**2 * (x_h**2 - a**2) / a**2))
    y2_hyperbola.append(-math.sqrt(b**2 * (x_h**2 - a**2) / a**2))

ax1.plot(x_circle, y1_circle, "r")
ax1.plot(x_circle, y2_circle, "r")
ax2.plot(x_ellipse, y1_ellipse, "g")
ax2.plot(x_ellipse, y2_ellipse, "g")
ax3.plot(x_hyperbola, y1_hyperbola, "b")
ax3.plot(x_hyperbola, y2_hyperbola, "b")
ax3.plot(-x_hyperbola, y1_hyperbola, "b")
ax3.plot(-x_hyperbola, y2_hyperbola, "b")
```

```python
# 3.4) уравнение 1
x = np.linspace(-2, 5, 201)
y1 = x**2 - 1
y2 = (np.exp(x) + x - 1) / x
plt.plot(x, y1)
plt.plot(x, y2)

def system(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x*(1-y) - 1)

x1, y1 =  fsolve(system, (5, 20))           
print (x1, y1)

result = []
for x_, y_ in zip(x, y):
    x1, y1 =  fsolve(system, (x_, y_))           
    result.append((round(x1, 2), round(y1)))
print(set(result))
# уравнение 2
```
